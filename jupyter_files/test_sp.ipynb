```python
import cv2
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# plt.style.use('ggplot')

%matplotlib inline

def sp_noise(image, prob=0.01):
    output = np.zeros(image.shape,np.uint8)
    thres = 1 - prob 
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            rdn = random.random()
            if rdn < prob:
                output[i][j] = 0
            elif rdn > thres:
                output[i][j] = 255
            else:
                output[i][j] = image[i][j]
    return output
```

```python
img = cv2.imread("../../Downloads/car_detection_sample1.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
out = sp_noise(img, 0.2)
plt.imshow(out)
```
